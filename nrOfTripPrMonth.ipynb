```python
#Take alot of time.. Use static list generated from this instead
nrOfTrips = []
for year in range(2016,2019): #FOR LEGACY LINKS (2016 - 2018)
    for i in range(1,13):
        try:
            month = str(i).zfill(2)
            URL = f"https://data-legacy.urbansharing.com/oslobysykkel.no/{year}/{month}.csv.zip"
            print(URL)
            df = pd.read_csv(URL)
            nrOfTrips.append(len(df.index))
        except:
            print(f'No data for {month}/{year}')
            nrOfTrips.append(0)
            
for year in range(2019,2021): #FOR CURRENT LINKS (2019 ->)
    for i in range(1,13):
        try:
            month = str(i).zfill(2) #Convert 1 to 01 and so on, to mach the url format
            URL = f"https://data.urbansharing.com/oslobysykkel.no/trips/v1/{year}/{month}.csv"
            print(URL)
            df = pd.read_csv(URL)
            nrOfTrips.append(len(df.index))
        except:
            nrOfTrips.append(0) #To fill the month where the stations are closed
            print(f'No data for {month}/{year}') 
```

```python
for nr in nrOfTrips:
    print(nr, end=',')
```
